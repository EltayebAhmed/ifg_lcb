```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
def plot_pdfs(mode="smaller", n=1000, num_bins=50):
    """
    Plots the probability density functions (PDFs) of two distributions: 
    smaller and larger, based on the specified mode.
    
    Parameters:
    - mode: str, either "smaller" or "larger", determines which distribution to plot
    - n: int, number of samples to generate
    - num_bins: int, number of bins for the histogram
    """
    
    rng = np.random.RandomState(32)
    a = rng.uniform(0, 1, n)
    # Generate random data
    b = []
    if mode == "smaller":
        for a_ in a:
            b.append(rng.uniform(0, a_))

        title_b = "b is smaller than a"
    elif mode == "larger":
        for a_ in a:
            b.append(rng.uniform(a_, 1))

        title_b = "b is larger than a"
    
    plt.hist(a, bins=num_bins, density=True, alpha=0.5, label='a', color='blue')
    plt.title("PDF of a")
    plt.figure()
    plt.hist(b, bins=num_bins, density=True, alpha=0.5, label='b', color='red')
    # Add a trend line to the histogram of b
    b_density, b_bins, _ = plt.hist(b, bins=num_bins, density=True, alpha=0.01, label='b', color='red')
    b_centers = 0.5 * (b_bins[1:] + b_bins[:-1])
    plt.plot(b_centers, b_density, label='Trend line', color='black')
    plt.legend()
    plt.title(title_b)

plot_pdfs("larger", n=10000)
```
